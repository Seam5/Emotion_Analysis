## 감성대화 말뭉치 기반 데이터 set 전처리

```python
import pandas as pd
import numpy as np
```

```python
data_t = pd.read_excel('/content/감성대화말뭉치(최종데이터)_Training.xlsx')
data_v = pd.read_excel('/content/감성대화말뭉치(최종데이터)_Validation.xlsx')
```

```python
data_v
```

```python
data_t = data_t.fillna("")
data_v = data_v.fillna("")
```

```python
# 컬럼 간 str 합치기
data_t["사람문장"] = data_t["사람문장1"].astype(str) + data_t["사람문장2"].astype(str) + data_t["사람문장3"].astype(str)
data_v["사람문장"] = data_v["사람문장1"].astype(str) + data_v["사람문장2"].astype(str) + data_v["사람문장3"].astype(str)
```

```python
df_concat = pd.concat([data_t, data_v])
```

```python
final_df = df_concat[["사람문장", "감정_대분류"]]
```

```python
final_df = final_df.rename({"감정_대분류" : "감정"}, axis = 1)
```

```python
# 컬럼 내 공백 제거
final_df["감정"] = final_df["감정"].apply(lambda x:x.strip())
```

```python
final_df["감정"].value_counts()
```

```python
final_df["감정"] = np.where(final_df["감정"].str.match("상처"), "슬픔", final_df["감정"])
```

```python
final_df["감정"].value_counts()
```

```python
final_df = final_df.reset_index(drop = True)
```

```python
final_df = final_df.rename({"사람문장" : "발화"}, axis = 1)
```
